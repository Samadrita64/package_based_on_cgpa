```python
#simple linear regression
#formula is y=mx+c
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('package.csv')
df
```

```python
df.head(2)
```

```python
df.tail()
```

```python
plt.scatter(df['CGPA'],df['PACKAGE(LPM)'],color='black')
plt.title("package prediction based on cgpa",color='green')
plt.xlabel("studentcgpa",color='red')
plt.ylabel("student package",color='red')
plt.show()
```

```python
x=df.iloc[:,:1]
y=df.iloc[:,-1]
print("all data of cgpa are:\n",x)
```

```python
print("all data of packages are:\n",y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression                               
```

```python
lr=LinearRegression()#creating object of linear regression
lr
```

```python
lr.fit(x_train,y_train)#best fit line draw
```

```python
x_test
```

```python
y_test
```

```python
lr.predict(x_test.iloc[1].values.reshape(1,1))
```

```python
plt.scatter(df['CGPA'],df['PACKAGE(LPM)'],color='black')
plt.plot(x_train,lr.predict(x_train),color='orange',linewidth=10)
plt.title("package prediction based on cgpa",color='green')
plt.xlabel("studentcgpa",color='red')
plt.ylabel("student package",color='red')
plt.show()
```

```python
m=lr.coef_
m
```

```python
c=lr.intercept_
c
```

```python
cgpa=float(input("enter your cgpa:-"))
package=m*cgpa+c
print("your predicted package based on your cgpa\n",package)
```
